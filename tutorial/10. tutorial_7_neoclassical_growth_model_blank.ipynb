## 1. Solow's growth model

#### - Solow Robert, 1956, "A contribution to the theory of economic growth," *Quarterly Journal of Economics*, 70(1), pp.65-94. 
#### - Basis for RBC and New Keynesian model
#### - Assume a one-sector producion technology in which output is a homogeneous good that can be consumed or invested.
#### - Investment is financed by saving. (exogenous saving rate and no household optimization)
#### - Capital accumulates with investments but depreciates with an arbitrary depreciation rate.

## 1-1. Setup

#### Production function is 
#### $$Y_t = a F(K_t, L_t)\quad,$$
#### where $Y_t$ is output, $a$ is the level of technology, $K_t$ is capital stock, and $L_t$ is labor.

#### Assumption of properties of $F(K_t, L_t)$
#### - twice continuously differentiable and strictly concave: $F_i >0$, $F_{ii} <0$, $F_{ij} >0$
#### - Inada condition: $F(K_t,0) = 0$, $F(0, N_t) = 0$, $F_N(K_t,0) = \infty$, $F_K(0,N_t) = \infty$, $F_N(K_t,\infty) = 0$, $F_K(\infty,N_t) = 0$

#### The population grows at a consant rate $\ell$:  $$L_{t+1} = (1 + \ell)L_{t}$$
#### Capital accumulates as follows,
#### $$K_{t+1} = (1-\delta)K_{t} + I_{t}\quad ,$$
#### where $I_t$ is the investment at time $t$ and it is financed by saving: $I_t = S_t$.
#### Saving is a portion of output:
#### $$S_t = s Y_t \quad,$$
#### where the parameter $s$ is exogenously given.
#### Output is either consumed or invested: $Y_t = C_t + I_t$

## 1-2. Capital accumulation

#### We can derive the following capital accumulation equation:
#### $$K_{t+1} = (1-\delta)K_{t} + s a F(K_t, L_t)\quad .$$

#### Divide this equation by $L_{t+1} = (1 + \ell)L_{t}$ , we get,
#### $$ k_{t+1} = \frac{(1-\delta)}{(1 + \ell)}k_t + \frac{s a F(K_t, L_t)}{(1 + \ell)L_{t}}\quad,$$
#### where capital per worker $k_{t} = \frac{K_{t}}{L_{t}}$.

#### We rewrite $F(K_t, L_t)$ as per worker term because of the assumption of homogenous of degree $1$:
#### $$\frac{F(K_t, L_t)}{L_t} = F(\frac{K_t}{L_t}, \frac{L_t}{L_t}) = F(k_t,1) \equiv f(k_t)$$

#### Then, the capital accumulation equation is rewritten as follows,
#### $$ k_{t+1} = \frac{(1-\delta)}{(1 + \ell)}k_t + \frac{s a }{(1 + \ell)}f(k_t)\quad,$$

#### Now, we assumue $f(k_t) = k_t^\alpha$ (decreasing returns to scale $0<\alpha<1$) and obtain the difference equation for the effective capital labor latio.
#### $$ k_{t+1} = \frac{(1-\delta)}{(1 + \ell)}k_t + \frac{s a}{(1 + \ell)} k_t^\alpha\quad,$$
